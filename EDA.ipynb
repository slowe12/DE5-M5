```python
# EDA.ipynb — Cell 1
from pathlib import Path
import pandas as pd

# Point to your project root (the folder that contains 'data')
# If EDA.ipynb is already in that folder, Path.cwd() is fine.
ROOT = Path.cwd()

DATA_DIR = ROOT / "data"

# File names (exactly as shown in Explorer)
books_csv = DATA_DIR / "03_Library Systembook.csv"
customers_csv = DATA_DIR / "03_Library SystemCustomers.csv"

books_csv, customers_csv
```

```python
def quick_check(df):
    print("Shape (rows, columns):", df.shape)
    print("\nColumn types:")
    print(df.dtypes)

    print("\nMissing values per column:")
    print(df.isna().sum())

    print("\nFirst 3 rows:")
    print(df.head(3))
```

```python
quick_check(books_df)
quick_check(cust_df)
```

```python
books_df_v01 = books_df.dropna()
books_df_v01.head()
```

```python

books_df_v01["Book checkout"] = (
    books_df_v01["Book checkout"]
    .astype(str)
    .str.replace('"', '', regex=False)   # remove quotes
    .str.strip()                          # remove spaces
    .pipe(pd.to_datetime, errors="coerce", dayfirst=True)
)

# Clean Book Returned
books_df_v01["Book Returned"] = (
    books_df_v01["Book Returned"]
    .astype(str)
    .str.replace('"', '', regex=False)
    .str.strip()
    .pipe(pd.to_datetime, errors="coerce", dayfirst=True)
)
```

```python
# Clean Book checkout
books_df_v01.head()
```

```python
# Convert "X weeks" → X*7 days
books_df_v01["Days allowed to borrow"] = (
    books_df_v01["Days allowed to borrow"]
        .str.split()
        .str[0]        # extract the number
        .astype(int)   # convert to int
        .mul(7)        # multiply by 7 days
)
books_df_v01.head()
```
